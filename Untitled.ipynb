```python
import pandas as pd
```

```python
df=pd.read_csv("recipes.csv")
df.head()
```

```python
import numpy as np

df.rename(columns={2: 'origin',3: 'destination'}, inplace=True) #for simplicity i took 2 columns since i have 10.000+ recipes but these steps can be proceed further
dfweight = df.groupby(['origin', 'destination']).size()


dfweight # edge weights between each node, for example zucchini-onion 3 but carrots only 1...
```

```python

links = dfweight.index.unique()
weights = dfweight.tolist()


#creating lists and weights for edge list
```

```python
input_tuple = [link + (weight,) for link, weight in zip (links, weights)]

input_tuple #our edge_list structure is ready for graph representationt
```

```python
input_tuple[:10]
```

```python
import matplotlib.pyplot as plt
plt.hist(weights,bins=max(weights))
plt.xlim(1,5)
plt.show()  #there is power distribution as you can see most of the nodes has degree 1
# the nodes such as salt has more weights since almost every food includes salt
```

```python
import networkx as nx
g_und = nx.Graph()
```

```python
g_und.name = 'Undirected Graph'
```

```python

llista1=np.unique(df.origin)
df.dropna(subset = ["destination"], inplace=True)
llista2=np.unique(df.destination)                            #USED UNIQUE TO PREVENT DATA RECURRENCE

ingnodes=np.unique(np.concatenate([llista1,llista2],axis=0)) #all nodes concenated

```

```python
g_und.add_nodes_from(ingnodes,bipartite=1) 
```

```python
g_und.add_weighted_edges_from(input_tuple) #we created bipartite unconnected graph
```

```python

nx.is_connected(g_und) #it is expected since we have 10.000+ recipes
```

```python
def plot_degree_distribution(wiki):
    degs = {}
    for n in wiki.nodes():
        deg = wiki.degree(n)

        if deg not in degs:
            degs[deg] = 0

        degs[deg] += 1
    items = sorted(degs.items())

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot([k for (k, v) in items], [v for (k, v) in items])
    ax.set_xscale('log')
    ax.set_yscale('log')
    plt.title("Degree Distribution")
    fig.savefig("degree_distribution.png")


    
plot_degree_distribution(g_und)
```
